```python
# Arbitry Arguments / Variable lenght argumnets

# 1. Non keyworded argument *args
# 2. Keywarded Arguments *kwargs
```

```python
def printdetails(*args):
    for i in args:
        print(i)

printdetails("Omkar","is","Good","Boy")
```

```python
def printdetails(args1,args2,*args):
    print("args1:",args1)
    print("args2:",args2)
    for i in args:
        print(i)

printdetails("Omkar","is","Good","Boy")
```

```python
def printdetails(**args):
    for i,j in args.items():
        print("%s=%s"%(i,j))

printdetails(k1="Omkar",k2="is",k3="Good",k4="Boy")
```

```python
def printdetails(a1,a2,**args):
    print(a1)
    print(a2)
    for i,j in args.items():
        print("%s=%s"%(i,j))

printdetails("Omkar","is",k3="Good",k4="Boy")
```

```python
def factno(no1):
    fact = 1
    for i in range(1, no1 + 1):
        fact = fact * i
    return fact

a = factno(10)
print(a)
```

```python
def areacircle(r):
    return 3.14*r*r

a=areacircle(5)
print(a)
```
