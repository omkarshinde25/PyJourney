```python
d1={}
d2={'1':"Apple",'2':"Mango"}
d3={'name':'Omkar','age':22,'City':'Pune'}
d4=dict(name='Ram',age=21,city="Pune")

print(d1)
print(d2)
print(d3)
print(d4)
```

```python
# how to access the emenet of dicitionary
```

```python
d2={'name':'Omkar','age':22,'city':'pune'}

print("Name :",d2['name'])
print("Age :",d2['age'])
print("City :",d2['city'])
```
