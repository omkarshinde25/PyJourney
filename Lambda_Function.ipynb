```python
# Lambda Function
```

```python
# Function which declare using lambda keywords, dont have a name
# these are anonymous function
# only one time executed not multiple time reuse

# lambda arguments:expression
#   \          \          \
```

```python
n=lambda x,y:x+y
print(n(10,20))
```

```python
m= lambda x,y:x*y

print(m(2,2))
```

```python
n=lambda x:3.14*x*x

print(n(3))
```

```python
no= lambda n:"Zero" if n==0 else "Even" if n%2==0 else "Old"

print(no(5))

print(no(10))

print(no(0))
```

```python
n=lambda x,y:(x+y,x-y,x*y,x/y,x%y)

res=n(10,2)
print(res)
```

```python
n=lambda x,y:(x+y,x-y,x*y,x/y,x%y)

print(n(10,2))
```

```python
num=[1,2,3,4,5,6]

is_even=filter(lambda a:a%2==0,num)

print(list(is_even))
```

```python
num=[1,2,3,4,-1,-2,4,5]

is_postive=filter(lambda a:a>0, num)

print(list(is_postive))
```

```python
# Map Function - map()
```

```python
num=[1,2,3,4,5,6]

square=map(lambda a:a**2,num)

print(list(square))
```
