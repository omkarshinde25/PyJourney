```python
# Object
# Object is a real world which having state and behaviour

# State- data about your object
# Behaviour - methods / functionality about the object

# there are two types of object
# 1) Physical Object - we can see we can touch feel eg- Dog
# 2) Virtual Object - we can't see and can't touch and feel eg - Bank Account
```

```python
# Class - class is a blueprint of object
# class is collection of varibale, methods and object
```

```python
class person:
    id=10
    name='Omkar'
    age=30

    def show_details(self):
        print("Id:",self.id, "Name:",self.name, "Age: ",self.age)

p1=person()
p1.show_details()
print(p1.id)
print(p1.name)
print(p1.age)
```

```python
class person: # class
    id=101
    name="Ram Shinde"
    age=21
    city="Pune"

    def show_details(self):
        print("Id",self.id,"Name",self.name,"Age",self.age,"City",self.city)

p1=person() # Object
p1.show_details()

p2=person()
p2.show_details() # Method
    
```

```python
# 1) Default Cunstroctor
# it is cunstroctor automatically executed on the background use for assign default values to the variables. 

# 2) Parameterized Cunstroctor
# A cunstroctor create 

# 3) Not Parameterized cunstroctor
```

```python
class Person:
    def __init__(self,id,name,age):
        self.id=id
        self.name=name
        self.age=age

    def show_details(self):
        print("Id:",self.id,"Name:",self.name,"Age:",self.age)

p1=Person(101,'Omkar',22)
p1.show_details()

p2=Person(102,'Ram',25)
p2.show_details()
```

```python
class person:
    def __init__(self):
        print("Non Parameterised Cunstrocutor Executed")

    def show_details(self,id,name,age):
        print("Id:",id,"Name:",name,"Age:",age)


p1=person()
p1.show_details(101,'Omkar',22)

p2=person()
p2.show_details(102,"Raj",23)
```

```python
class animal:
    name='Sunny'
    def make_sound(self):
        print("Making Sound")

class dog(animal):
    def bark(self):
        print("Dog is Braking")

d1=dog()
d1.bark() # method access of own class
d1.make_sound() # Method access of another class
print(d1.name) # Variable access of another class
```
