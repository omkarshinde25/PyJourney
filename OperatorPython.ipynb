```python
# Membership Operator - in, not in
```

```python
l1=[12,23,3,44,55,66,67]
```

```python
x=12
```

```python
print(x in l1)
```

```python
y= 98
```

```python
print(y not in l1)
```

```python
print("Original",l1)
if x in l1:
    print("12 is part of l1")
else:
    print("12 is not part of l1")

if y not in l1:
    print("98 is not part of l1")
else:
    print("98 is part of l1")
```
