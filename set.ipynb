```python
# mutable object

# add() - add a single element into the set

# update () - allow to add multiple elemenets
```

```python
s1={'SQL','Power BI','Excel'}

print("Original set :", s1)

s1.add("Python")

print("Set after add",s1)

s1.update(["ML","DL","AI"])

print("Multiple add in set",s1)
```

```python
s1={1,2,3,4,6,7}
s2={5,6,7,8,9,10}

print("s1:",s1)

print("s2:",s2)

print("Union of s1|s2 :", s1|s2)

print("Union of s1.union(s2):",s1.union(s2))
```

```python
s1={1,2,3,4}
s2={2,3,4,5,6}

print("s1",s1)
print("s2",s2)

print("Intersection of s1&s2",s1&s2)

print("Intersection of s1.intersection(s2)",s1.intersection(s2))
```

```python
# Differenece between set 

# differnce() and - 
```

```python
s1={1,2,3,4,5}

s2={4,5,6,7,8}

print("s1:",s1)

print("s2:",s2)

print("Difference of s1-s2",s1-s2)

print("Difference of s1.difference(s2)",s1.difference(s2))
```
