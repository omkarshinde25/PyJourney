```python
for i in range(1,5):
    for j in range(1,5):
        print("*",end=' ')
    print()
```

```python
for i in range(1,6):
    for j in range(1,6):
        print("*",end=' ')
    print()
```

```python
for i in range(1,5):
    for j in range(1,5):
        if j<=i:
            print("*",end=' ')
        else:
            print(' ', end=' ')
    print()
```

```python
for i in range(1,5):
    for j in range(1,5):
        if j>=5-i:
            print("*", end=' ')
        else:
            print(" ", end =' ')
    print()
```
